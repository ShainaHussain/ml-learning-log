# Regression Example

This example shows how to predict the future value of salary

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### load the data from data source

```python
df = pd.read_csv('Salary_Data.csv')
print(df.head())
```

#### exploratory data analysis

explore the data to find out the patterns which can be used for prediction of salary

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

#### Data Visualization

```python
plt.scatter(df['YearsExperience'], df['Salary'], label="Salary")
plt.title('Salary vs Experience')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.tight_layout()
```

```python
sns.pairplot(df)
```

#### Decide the dependent and independent variables

```python
# independent variables
x = df.drop('Salary', axis=1)

# dependent variable
y = df['Salary']
```

#### create a regression model

```python
from sklearn.linear_model import LinearRegression

# create a model
model = LinearRegression()

# train the model
model.fit(x, y)
```

#### prediction

find the salary for an employee having 5.4 years of experience

```python
salaries = model.predict([[5.4]])
print(salaries[0])
```

#### result visualization

this section will show how to draw the best fit regression line

```python
# observed values
plt.scatter(df['YearsExperience'], df['Salary'], label="Salary")

# mean line
means = np.ones(df['YearsExperience'].size) * df['Salary'].mean()
plt.plot(df['YearsExperience'], means, label="Mean", color="green")
plt.scatter(df['YearsExperience'], means, color="green")

# regression line
predicted_salaries = model.predict(x)
plt.plot(df['YearsExperience'], predicted_salaries, label="Regression Line", color="red")
plt.scatter(df['YearsExperience'], predicted_salaries, color="red")

plt.title('Salary vs Experience')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.tight_layout()
plt.show()
```
