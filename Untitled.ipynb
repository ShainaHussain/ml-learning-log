```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , LogisticRegression
from sklearn.neighbors import KNeighborsRegressor , KNeighborsClassifier
from sklearn.metrics import mean_squared_error , accuracy_score
```

```python
np.random.seed(42)
n =200
TV =np.random.randint(50,300,n)
Radio = np.random.randint(10,100,n)
Newspaper = np.random.randint(0,50,n)
```

```python
Sales = 3+ 0.05*TV + 0.1*Radio + 0.02*Newspaper + np.random.normal(0,2,n)

df = pd.DataFrame({
    "TV":TV,
    "Radio":Radio,
    "Newspaper":Newspaper,
    "Sales":Sales
    })
print(df.head())
```

```python
X = df[["TV","Radio","Newspaper"]]
y = df["Sales"]
```

```python
sns.pairplot(df,x_vars=["TV","Radio","Newspaper"],y_vars="Sales",height =4,aspect =1,kind ="scatter")
plt.suptitle("Sales vs Advertisemnet Budget",y=1.02)
plt.show()
```

```python
X_train ,X_test ,y_train ,y_test =train_test_split(X,y,test_size=0.3 ,random_state =42)
```

```python
lin_reg =LinearRegression()
lin_reg.fit(X_train ,y_train)
y_pred_train = lin_reg.predict(X_train)
y_pred_test =lin_reg.predict(X_test)

print("Training MSE (Linear Regression:",mean_squared_error(y_train , y_pred_train))
print("Test MSE:",mean_squared_error(y_test,y_pred_test))
```

```python
# knn regression
train_errors =[]
test_errors =[]
k_values = range(1,21)
for k in k_values:
    knn =KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train ,y_train)
    y_train_pred = knn.predict(X_train)
    y_test_pred = knn.predict(X_test)

    train_errors.append(mean_squared_error(y_train , y_train_pred))
    test_errors.append(mean_squared_error(y_test , y_test_pred))

plt.plot(k_values,train_errors,label = "Training MSE")
plt.plot(k_values,test_errors,label = "Test MSE")
plt.xlabel("K (Number of Neighbors)")
plt.ylabel("MSE")
plt.title("Model Selection - KNN Regressor")
plt.legend()
plt.show()
```

```python
## bias variance regression
best_k = k_values[np.argmin(test_errors)]
print("Best K value (min test MSE):" ,best_k)
```

```python
# classification task
df["High_Sales"] = (df["Sales"]>20).astype(int)
X_class = df[["TV","Radio","Newspaper"]]
y_class =df["High_Sales"]
X_train_c , X_test_c , y_train_c , y_test_c = train_test_split(X_class,y_class,test_size =0.3,random_state=42)
```

```python
# logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train_c , y_train_c)
y_pred_log = log_reg.predict(X_test_c)
print("Logistic Regression Accuracy:",accuracy_score(y_test_c,y_pred_log))
```

```python
## KNN classifier
knn_c = KNeighborsClassifier(n_neighbors =best_k)
knn_c.fit(X_train_c,y_train_c)
y_pred_knn = knn_c.predict(X_test_c)
print("KNN classifier Accuracy:",accuracy_score(y_test_c,y_pred_knn))
```

```python

# -------------------------
# 10. Business Insights
# -------------------------
print("Linear Regression Coefficients (TV, Radio, Newspaper):", lin_reg.coef_)
print("Intercept:", lin_reg.intercept_)

print("\nBusiness Insight:")
print("1. TV has the strongest impact on sales (highest coefficient).")
print("2. Increasing Radio budget by $10,000 increases sales by ~", lin_reg.coef_[1]*10, "units.")
print("3. KNN worked best with K =", best_k, "neighbors.")
```
