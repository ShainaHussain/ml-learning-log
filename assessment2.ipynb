```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data = pd.read_csv("exam_assessment_questions_20250925155000.csv")
```

```python
X = data[['SAT']]  
y = data['GPA']    
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, y, color="blue", label="Observed values")
plt.axhline(y.mean(), color="green", linestyle="--", label="Mean line")
plt.plot(X, y_pred, color="red", label="Best-fit regression line")
plt.xlabel("SAT")
plt.ylabel("GPA")
plt.title("SAT vs GPA with Regression Line")
plt.legend()
plt.show()
```

```python
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
sat_score = np.array([[2000]])
predicted_gpa = model.predict(sat_score)[0]
print("Predicted GPA for SAT=2000:", predicted_gpa)
```
